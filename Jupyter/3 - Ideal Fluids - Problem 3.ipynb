```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np

#rc('text', usetex=True)
#rc('font',**{'family':'serif','serif':['Palatino']}, size=18)

r,w = np.loadtxt("wind_speed.csv", delimiter=",", unpack=True)

plt.plot(r, w, "o")

a = 32.0
Omega = 2.5

r = np.linspace(0, r[-1], 1000)
w_rv = np.zeros(1000)
w_rv[r<a] = Omega * r[r<a]
w_rv[r>a] = Omega * a**2 / r[r>a]

plt.plot(r, w_rv)
```
