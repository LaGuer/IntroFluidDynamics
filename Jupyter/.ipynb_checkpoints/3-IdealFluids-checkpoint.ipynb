## Figures and Examples

Before we start any actual plotting code, here's some useful things to incude first.

```python
%matplotlib inline

import matplotlib
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
import sympy as sp

rc('text.latex',preamble='\\usepackage{libertine}\n\\usepackage[libertine]{newtxmath}')
rc('font',**{'family':'serif','serif':['Linux Libertine O']}, size=18)
rc('text', usetex=True)
```

## Figure 3.3:  Uniformly rotating fluid

```python
Omega = 1.0
g = 9.8

def fun(x, y):
  return 0.5 * Omega / g * (x**2 + y**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-3.0, 3.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array([fun(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z, color="lightblue")

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')

plt.savefig("fig_paraboloid.pdf")
```

## Figure 3.5: Streamlines for flow past a cylinder

This code calculates and plots streamlines; however, be aware that it does so using the stream function, which we won't cover until later.

```python

x = sp.Symbol('x', real=True)
y = sp.Symbol('y', real=True)
z = sp.Symbol('z', complex=True)

z = x + sp.I*y

U = 0.4
a = 1.0
Gamma = 0.0
alpha = 0.0

w = U * (z * sp.exp(-sp.I * alpha) + a*a / (z * sp.exp(-sp.I * alpha))) - sp.I * Gamma / (2 * sp.pi) * sp.ln(z)

psi = sp.im(w)
#psi = sp.Abs((z-d)/(z+d))
#psi = sp.Abs(z-d)

psi = sp.lambdify((x, y), psi, 'numpy')

xx = np.linspace(-3.0, 3.0, 100)
yy = np.linspace(-2.0, 2.0, 100)
xx, yy = np.meshgrid(xx, yy)

psi(xx, yy)[:100//2, :100//2] = np.ma.masked

fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(1, 1, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.axes().set_aspect('equal')
plt.axis('off')
levs = np.linspace(-0.5, 0.5, 11)
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
plt.contour(xx, yy, psi(xx, yy), colors='black', levels=levs)
circ = plt.Circle((0,0),radius=a,fc='lightgrey', fill=True, alpha=1, zorder=2, color="black")
ax.add_patch(circ)

ax.add_patch(plt.Circle((0,1),radius=0.04,fc='black', fill=True, alpha=1, zorder=2, color="black"))
plt.arrow(0, 1, 1, 0, color="red", lw=3, head_width=0.15, zorder=3)

ax.add_patch(plt.Circle((0,-1),radius=0.04,fc='black', fill=True, alpha=1, zorder=2, color="black"))
plt.arrow(0, -1, 1, 0, color="red", lw=3, head_width=0.15, zorder=3)

ax.add_patch(plt.Circle((1,0),radius=0.04,fc='black', fill=True, alpha=1, zorder=2, color="black"))
ax.add_patch(plt.Circle((-1,0),radius=0.04,fc='black', fill=True, alpha=1, zorder=2, color="black"))

plt.savefig("fig_cylinder_stream.pdf", bbox_inches='tight')
```

## Figure 3.6: Pressure

```python
def p(x, y):
    r2 = x**2 + y**2
    costheta2 = x**2 / r2
    # the 3.01 is here to make sure there's no negative values of pressure
    # the sqrt is there to scale the pressure to make the final image better
    return np.sqrt(3.01 + (4*x**2 - 2.0 * r2 - 1)/r2/r2)

x = np.linspace(-3, 3, 1000)
y = np.linspace(-2, 2, 1000)

X, Y = np.meshgrid(x, y)
Z = p(X, Y)

fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(1, 1, 1)

plt.axes().set_aspect('equal')

plt.imshow(Z, extent=[-3, 3, -2, 2], origin='lower',
           cmap='gist_rainbow', alpha=1)
#contours = plt.contour(X, Y, Z, 30, colors='grey', alpha=0.5)
plt.axis('off')
circ = plt.Circle((0,0),radius=a,fc='lightgrey', fill=True, alpha=1, zorder=3, color="black")
ax.add_patch(circ)

plt.contour(xx, yy, psi(xx, yy), colors='white', levels=levs, alpha=0.5)

plt.savefig("fig_cylinder_pressure.pdf", bbox_inches='tight')
```
