# Two Vortexes

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

x = sp.Symbol('x', real=True)
y = sp.Symbol('y', real=True)
z = sp.Symbol('z', complex=True)

z = x + sp.I*y

d = 1.0
y0 = 0.0
z0 = d + sp.I*y0
z0s= d - sp.I*y0
Gamma = 1.0
U = Gamma/(4*sp.pi*d)

w =  -sp.I * U * z - sp.I * Gamma / 2 / sp.pi * sp.log(z - z0) + sp.I * Gamma / 2 / sp.pi * sp.log(z + z0s)

psi = sp.im(w)

psi = sp.lambdify((x, y), psi, 'numpy')

xx = np.linspace(-4.0, 4.0, 100)
yy = np.linspace(-4.0, 4.0, 100)
xx, yy = np.meshgrid(xx, yy)


fig = plt.figure(figsize=(12, 6), dpi=80)
ax = fig.add_subplot(1, 1, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.axes().set_aspect('equal')
plt.axis('off')
levs = np.linspace(-0.5, 0.5, 21)
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
plt.contour(xx, yy, psi(xx, yy), colors='black', levels=levs)
plt.savefig("fig_vortex_pair2.pdf")
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

x = sp.Symbol('x', real=True)
y = sp.Symbol('y', real=True)
z = sp.Symbol('z', complex=True)

z = x + sp.I*y

a = 3.0
b = 0.5
Gamma = 1.0
U = 0.2
w = U*z -1.0 * sp.I * Gamma / 2 / sp.pi * sp.ln(sp.sin(sp.pi * z / a)) + 1.0 * sp.I * Gamma / 2 / sp.pi * sp.ln(sp.sin(sp.pi * (z - sp.I*b + 0.5 * a) / a))
print(Gamma / 2.0 / a * np.tanh(np.pi*b/a))
psi = sp.im(w)

psi = sp.lambdify((x, y), psi, 'numpy')

xx = np.linspace(-4.0, 4.0, 1000)
yy = np.linspace(-2.0, 2.5, 1000)
xx, yy = np.meshgrid(xx, yy)


fig = plt.figure(figsize=(12, 6), dpi=80)
ax = fig.add_subplot(1, 1, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.axes().set_aspect('equal')
plt.axis('off')
levs = np.linspace(-0.5, 0.5, 31)
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
plt.contour(xx, yy, psi(xx, yy), colors='black', levels=levs)
#plt.savefig("fig_vortex_pair2.pdf")
```
