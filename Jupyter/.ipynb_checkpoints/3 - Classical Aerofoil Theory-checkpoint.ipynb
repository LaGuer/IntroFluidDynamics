# Example:  Method of Images 1

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

x = sp.Symbol('x', real=True)
y = sp.Symbol('y', real=True)
z = sp.Symbol('z', complex=True)

z = x + sp.I*y

d = 1

psi = sp.Abs((z-d)/(z+d))


u = sp.lambdify((x, y), psi.diff(y), 'numpy')
v = sp.lambdify((x, y), -psi.diff(x), 'numpy')

Y,X = np.ogrid[-2:2:0.15, 0:4:0.15]

print X, Y

uu = u(X, Y)
vv = v(X, Y)

plt.figure(figsize=(6, 6), dpi=80)
plt.xlabel('x')
plt.ylabel('y')
plt.axes().set_aspect('equal')
plt.streamplot(X, Y, uu, vv, density=1)
plt.show()
```

# Example:  Method of Images 2

I'd like the streamlines to connect and don't particularly care if the arrows are there, so I'll use contour lines instead.

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

x = sp.Symbol('x', real=True)
y = sp.Symbol('y', real=True)
z = sp.Symbol('z', complex=True)

z = x + sp.I*y

d = 1
Gamma = 1

w = -sp.I * Gamma / (2*sp.pi) * sp.log(z-d) + sp.I * Gamma / (2*sp.pi) * sp.log(z+d)

psi = sp.im(w)
#psi = sp.Abs((z-d)/(z+d))
#psi = sp.Abs(z-d)

psi = sp.lambdify((x, y), psi, 'numpy')

xx = np.linspace(0.0, 8.0, 100)
yy = np.linspace(-4.0, 4.0, 100)
xx, yy = np.meshgrid(xx, yy)

plt.figure(figsize=(6, 6), dpi=80)
plt.xlabel('x')
plt.ylabel('y')
plt.axes().set_aspect('equal')
plt.axis('off')
levs = np.linspace(-0.5, 0.5, 20)
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
plt.contour(xx, yy, psi(xx, yy), colors='black', levels=levs)
plt.plot([0,0], [-4,4], color='black', linestyle='solid', linewidth=5)
```

# Making a circular boundary

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

x = sp.Symbol('x', real=True)
y = sp.Symbol('y', real=True)
z = sp.Symbol('z', complex=True)

z = x + sp.I*y

d = 1
Gamma = 1

w = -sp.I * Gamma / (2*sp.pi) * sp.log(z-d) + sp.I * Gamma / (2*sp.pi) * sp.log(z+d)

psi = sp.im(w)

psi = sp.lambdify((x, y), psi, 'numpy')

xx = np.linspace(-2.0, 6.0, 100)
yy = np.linspace(-4.0, 4.0, 100)
xx, yy = np.meshgrid(xx, yy)

fig = plt.figure(figsize=(6, 6), dpi=80)
ax = fig.add_subplot(1, 1, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.axes().set_aspect('equal')
plt.axis('off')
levs = np.linspace(-0.5, 0.5, 10)
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
plt.contour(xx, yy, psi(xx, yy), colors='black', levels=levs)
plt.plot([0,0], [-4,4], color='black', linestyle='solid', linewidth=1)
plt.plot([-2,6], [0,0], color='black', linestyle='solid', linewidth=1)
plt.plot(2.95,0, marker="o", color="black")
circ = plt.Circle((0,0),radius=1,color='black', linestyle="dashed", fill=False, )
ax.add_patch(circ)
plt.savefig("fig_circ_boundary.pdf")
```

# Streamlines in a circular boundary

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

x = sp.Symbol('x', real=True)
y = sp.Symbol('y', real=True)
z = sp.Symbol('z', complex=True)

z = x + sp.I*y

c = 0.4
a = 1
Gamma = 1

w = -sp.I * Gamma / (2*sp.pi) * sp.log(z-c)+ sp.I * Gamma / (2*sp.pi) * sp.log(z-a**2/c)

psi = sp.im(w)

psi = sp.lambdify((x, y), psi, 'numpy')

xx = np.linspace(-1.1, 1.1, 100)
yy = np.linspace(-1.1, 1.1, 100)
xx, yy = np.meshgrid(xx, yy)

fig = plt.figure(figsize=(6, 6), dpi=80)
ax = fig.add_subplot(1, 1, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.axes().set_aspect('equal')
#plt.axis('off')
levs = np.linspace(0.16, 1, 10)
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
plt.contour(xx, yy, psi(xx, yy), colors='black', levels=[0.16, 0.2, 0.3, 0.5, 0.4])
plt.plot(c,0, marker="o", color="black")
circ = plt.Circle((0,0),radius=a,color='black', fill=False, linewidth=5)
ax.add_patch(circ)
#plt.savefig("fig_circ_boundary3.pdf")
```

# Streamlines around a circular cylinder

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

x = sp.Symbol('x', real=True)
y = sp.Symbol('y', real=True)
z = sp.Symbol('z', complex=True)

z = x + sp.I*y

U = 0.4
a = 1.0
Gamma = 0.0
alpha = 0.0

w = U * (z * sp.exp(-sp.I * alpha) + a*a / (z * sp.exp(-sp.I * alpha))) - sp.I * Gamma / (2 * sp.pi) * sp.ln(z)

psi = sp.im(w)
#psi = sp.Abs((z-d)/(z+d))
#psi = sp.Abs(z-d)

psi = sp.lambdify((x, y), psi, 'numpy')

xx = np.linspace(-3.0, 3.0, 100)
yy = np.linspace(-2.0, 2.0, 100)
xx, yy = np.meshgrid(xx, yy)

psi(xx, yy)[:100//2, :100//2] = np.ma.masked

fig = plt.figure(figsize=(12, 6), dpi=80)
ax = fig.add_subplot(1, 1, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.axes().set_aspect('equal')
plt.axis('off')
levs = np.linspace(-0.5, 0.5, 10)
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
plt.contour(xx, yy, psi(xx, yy), colors='black', levels=levs)
circ = plt.Circle((0,0),radius=a,fc='black', fill=True, alpha=1)
#ax.add_patch(circ)
#plt.savefig("fig_cylinder1.pdf")
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

x = sp.Symbol('x', real=True)
y = sp.Symbol('y', real=True)
z = sp.Symbol('z', complex=True)

z = x + sp.I*y

def get_psi(U, a, Gamma, alpha):
        w = U * (z * sp.exp(-sp.I * alpha) + a*a / (z * sp.exp(-sp.I * alpha))) - sp.I * Gamma / (2 * sp.pi) * sp.ln(z)
        psi = sp.im(w)
        psi = sp.lambdify((x, y), psi, 'numpy')
        xx = np.linspace(-3.0, 3.0, 100)
        yy = np.linspace(-2.1, 2.1, 100)
        xx, yy = np.meshgrid(xx, yy)
        
        return xx, yy, psi
        

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row', figsize=(12,6))

plt.tight_layout()

ax1.axis('off')
ax1.set_aspect('equal')
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
xx, yy, psi = get_psi(0.4, 1.0, 0.0, 0.0)
ax1.contour(xx, yy, psi(xx, yy), colors='black', levels=[-0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6])
circ = plt.Circle((0,0),radius=a,fc='black', fill=True, alpha=1)
ax1.add_patch(circ)
ax1.text(0, -2.5, "(a)", ha="center", family="serif", size=16)


ax2.axis('off')
ax2.set_aspect('equal')
xx, yy, psi = get_psi(0.4, 1.0, -4, 0.0)
ax2.contour(xx, yy, psi(xx, yy), colors='black', levels=[ -0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6])
circ = plt.Circle((0,0),radius=a,fc='black', fill=True, alpha=1)
ax2.add_patch(circ)
ax2.text(0, -2.5, "(b)", ha="center", family="serif", size=16)

ax3.axis('off')
ax3.set_aspect('equal')
xx, yy, psi = get_psi(0.4, 1.0, -5.026, 0.0)
ax3.contour(xx, yy, psi(xx, yy), colors='black', levels=[ -0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6])
circ = plt.Circle((0,0),radius=a,fc='black', fill=True, alpha=1)
ax3.add_patch(circ)
ax3.text(0, -2.5, "(c)", ha="center", family="serif", size=16)

ax4.axis('off')
ax4.set_aspect('equal')
xx, yy, psi = get_psi(0.4, 1.0, -6.4, 0.0)
ax4.contour(xx, yy, psi(xx, yy), colors='black', levels=[ -0.7, -0.4, -0.3, 0, 0.1,  0.3, 0.5, 0.7])
circ = plt.Circle((0,0),radius=a,fc='black', fill=True, alpha=1)
ax4.add_patch(circ)
ax4.text(0, -2.5, "(d)", ha="center", family="serif", size=16)

plt.savefig("fig_cylinder2.pdf")
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

x = sp.Symbol('x', real=True)
y = sp.Symbol('y', real=True)
z = sp.Symbol('z', complex=True)
theta = sp.Symbol('theta', real=True)

z = x + sp.I*y

U = 0.4
c = 0.99
a = 1.0
l = 0.1

alpha = 0.5235
Gamma = -4*sp.pi*U*a*sp.sin(alpha)

zz = 0.5*z + sp.sqrt(0.25*z**2 - c**2)

w = U * ((zz+l) * sp.exp(-sp.I * alpha) + (a+l)**2 / ((zz+l) * sp.exp(-sp.I * alpha))) - sp.I * Gamma / (2 * sp.pi) * sp.ln(zz+l)

psi = sp.im(w)

psi = sp.lambdify((x, y), psi, 'numpy')

xx = np.linspace(0.01, 3.0, 100)
yy = np.linspace(-2.0, 2.0, 100)
xx, yy = np.meshgrid(xx, yy)

fig = plt.figure(figsize=(12, 6), dpi=80)
ax = fig.add_subplot(1, 1, 1)
plt.axes().set_aspect('equal')
plt.axis('off')
levs = np.linspace(-0.5, 0.5, 11)
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
plt.contour(xx, yy, psi(xx, yy), colors='black', levels=levs)
circ = matplotlib.patches.Ellipse((0,0), width = 2*(a+c**2/a), height = 2*(a-c**2/a), color='black')
#ax.add_patch(circ)

zz = 0.5*z - sp.sqrt(0.25*z**2 - c**2)

w = U * ((zz+l) * sp.exp(-sp.I * alpha) + (a+l)**2 / ((zz+l) * sp.exp(-sp.I * alpha))) - sp.I * Gamma / (2 * sp.pi) * sp.ln(zz+l)

psi = sp.im(w)

psi = sp.lambdify((x, y), psi, 'numpy')

xx = np.linspace(-3.0, -0.001, 100)
yy = np.linspace(-2.0, 2.0, 100)
xx, yy = np.meshgrid(xx, yy)

plt.axes().set_aspect('equal')
plt.axis('off')
levs = np.linspace(-0.5, 0.5, 11)
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
plt.contour(xx, yy, psi(xx, yy), colors='black', levels=levs)
circ = matplotlib.patches.Ellipse((0,0), width = 2*(a+c**2/a), height = 2*(a-c**2/a), color='black')
#ax.add_patch(circ)

z = (a+l) * sp.exp(sp.I*theta) - l

Z = z + c**2/z

x = sp.re(Z)
y = sp.im(Z)

xx = sp.lambdify(theta, x, "numpy")
yy = sp.lambdify(theta, y, "numpy")

thetas = np.linspace(0.0, 2.0*np.pi, 100)
#ax.plot(xx(thetas), yy(thetas), color="black")
ax.fill(xx(thetas), yy(thetas), color="black")
plt.savefig("fig_aero2.pdf")
```
